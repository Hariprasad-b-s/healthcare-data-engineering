```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```python
try:
    spark.stop()
except:
    pass
```

```python
spark = (SparkSession.builder.appName("HealthcareDataProcessing_DimDate")
    .master("local[*]")
    .config("spark.driver.memory", "4g")
    .config("spark.memory.offHeap.enabled", "true")
    .config("spark.memory.offHeap.size", "2g")
.getOrCreate())
```

```python
silver_dim_date_path = "../../data_lake/silver/dim_date"
gold_dim_date_path = "../../data_lake/gold/dim_date"
```

```python
df_dim_date = spark.read.parquet(silver_dim_date_path).withColumn("gold_timestamp", F.current_timestamp()).drop("silver_timestamp")
```

```python
df_dim_date.write.mode("overwrite").format("parquet").save(gold_dim_date_path)
```

```python
spark.stop()
```
